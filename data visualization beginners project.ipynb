# PLOTTING WITH MATPLOTLIB

# lineplot

```python
lineplot=df.plot(x='job_title' ,y='salary' ,kind='line' , linewidth=2 , figsize=(21,6),color='blue')
plt.title('salary VS Job',color='blue')
plt.xlabel('JOBS TITLE',color='blue')
plt.ylabel('SALARY',color='blue')
```


# AREA PLOT

```python
area_df=df[['work_year','salary_in_usd','job_title']]
area_df.head(100)
area_df
```

```python
areaplot=area_df.plot(kind='area', figsize=(19,7))
plt.title('SAlARY VS WORK YEAR',color='green')
plt.xlabel('WORK_YEAR  ',color='green')
plt.ylabel('SALARY',color='green')
```

# histogram

```python
hist_df=df[['work_year','salary']]
hist_df=hist_df.head(100)
hist_df
```

```python
histo=hist_df.plot(kind='hist', figsize=(15,7), bins=10)     
```
